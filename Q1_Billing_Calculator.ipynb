# 🛒 Question 1: Billing Calculator with Discount

**Concepts Covered:** User input, arithmetic operations, conditional statements, formatted output

**Scenario:** Uncle wants a tool where he can input items purchased by a customer and instantly calculate the total bill. A **5% discount** is applied if the bill exceeds ₹500.

````python
# ============================================================
#  Kirana Store - Billing Calculator with Discount
#  Assignment 1 | Question 1
# ============================================================

print("=" * 45)
print("       🏪 KIRANA STORE - BILL GENERATOR")
print("=" * 45)

items = []  # Will store (name, quantity, price_per_unit) tuples

# --- Input: 3 items ---
for i in range(1, 4):
    print(f"\n--- Item {i} ---")
    name     = input(f"  Enter item {i} name     : ").strip().title()
    quantity = int(input(f"  Enter quantity        : "))
    price    = float(input(f"  Enter price per unit (₹): "))
    items.append((name, quantity, price))

# --- Calculation ---
print("\n" + "=" * 45)
print("              📋 BILL SUMMARY")
print("=" * 45)
print(f"{'Item':<15} {'Qty':>5} {'Rate':>8} {'Amount':>10}")
print("-" * 45)

subtotal = 0
for name, qty, price in items:
    amount    = qty * price
    subtotal += amount
    print(f"{name:<15} {qty:>5} {price:>8.2f} {amount:>10.2f}")

print("-" * 45)
print(f"{'Subtotal':<30} ₹{subtotal:>10.2f}")

# --- Discount Logic ---
DISCOUNT_THRESHOLD = 500
DISCOUNT_RATE      = 0.05

if subtotal > DISCOUNT_THRESHOLD:
    discount   = subtotal * DISCOUNT_RATE
    final_bill = subtotal - discount
    print(f"{'Discount (5%)':<30} ₹{discount:>10.2f}")
    print("=" * 45)
    print(f"{'✅ Final Payable Amount':<30} ₹{final_bill:>10.2f}")
    print("\n🎉 Lucky! You saved ₹{:.2f} with a 5% discount!".format(discount))
else:
    print("=" * 45)
    print(f"{'✅ Final Payable Amount':<30} ₹{subtotal:>10.2f}")
    print("\n💡 Tip: Spend above ₹500 to get a 5% discount!")

print("=" * 45)
print("       Thank you for shopping with us! 🙏")
print("=" * 45)
````
