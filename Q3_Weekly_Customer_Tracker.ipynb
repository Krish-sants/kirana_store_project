# 📅 Question 3: Weekly Customer Tracker

**Concepts Covered:** Lists, loops, indexing, basic analytics

**Scenario:** Uncle records daily customer footfall and wants to analyze which days are busier, compare alternate-day totals, and find peak days.

````python
# ============================================================
#  Kirana Store - Weekly Customer Tracker
#  Assignment 1 | Question 3
# ============================================================

days = ["Monday", "Tuesday", "Wednesday", "Thursday",
        "Friday", "Saturday", "Sunday"]

print("=" * 45)
print("    📅 KIRANA STORE - WEEKLY FOOTFALL")
print("=" * 45)

# --- Input: customer count per day ---
footfall = []
for day in days:
    count = int(input(f"  Customers on {day:<12}: "))
    footfall.append(count)

# --- Analysis ---
# Alternate days: Mon(0), Wed(2), Fri(4), Sun(6)
group_a      = footfall[0:7:2]   # Mon, Wed, Fri, Sun
group_a_days = days[0:7:2]

# Remaining days: Tue(1), Thu(3), Sat(5)
group_b      = footfall[1:7:2]   # Tue, Thu, Sat
group_b_days = days[1:7:2]

total_a    = sum(group_a)
total_b    = sum(group_b)
grand_total = sum(footfall)
peak_day    = days[footfall.index(max(footfall))]
avg_daily   = grand_total / 7

# --- Display Report ---
print("\n" + "=" * 45)
print("           📊 WEEKLY REPORT")
print("=" * 45)
print(f"{'Day':<12} {'Customers':>10} {'Bar Chart'}")
print("-" * 45)
for day, count in zip(days, footfall):
    bar = "█" * (count // 5)   # 1 block per 5 customers
    print(f"{day:<12} {count:>10}   {bar}")

print("=" * 45)
print(f"\n📌 Group A ({', '.join(group_a_days)})")
print(f"   Total Customers : {total_a}")

print(f"\n📌 Group B ({', '.join(group_b_days)})")
print(f"   Total Customers : {total_b}")

print(f"\n📦 Total Weekly Customers : {grand_total}")
print(f"📈 Daily Average          : {avg_daily:.1f}")
print(f"🏆 Busiest Day            : {peak_day} ({max(footfall)} customers)")

if total_a > total_b:
    print("\n💡 Insight: Alternate days (Mon/Wed/Fri/Sun) are busier. Consider extra staff on those days.")
else:
    print("\n💡 Insight: Mid-week days (Tue/Thu/Sat) are busier. Consider running offers on those days.")
````
