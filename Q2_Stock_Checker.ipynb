# 📦 Question 2: Stock Availability Checker

**Concepts Covered:** Lists, string matching, input/output

**Scenario:** Customers often ask if certain items are in stock. Uncle needs a quick way to check that using a predefined stock list.

````python
# ============================================================
#  Kirana Store - Stock Availability Checker
#  Assignment 1 | Question 2
# ============================================================

# Predefined stock list
stock = ["sugar", "rice", "salt", "tea", "coffee",
         "dal", "atta", "oil", "soap", "toothpaste"]

print("=" * 45)
print("     📦 KIRANA STORE - STOCK CHECKER")
print("=" * 45)
print("Available stock items:")
for idx, item in enumerate(stock):
    print(f"  [{idx}] {item.title()}")
print("=" * 45)

# --- Keep checking until user types 'exit' ---
while True:
    query = input("\nEnter item name to check (or 'exit' to quit): ").strip().lower()

    if query == 'exit':
        print("\n👋 Thank you! Come again.")
        break

    if query in stock:
        position = stock.index(query)
        print(f"  ✅ '{query.title()}' is AVAILABLE in stock!")
        print(f"  📍 Shelf Index: {position} (Row {position // 5 + 1}, Slot {position % 5 + 1})")
    else:
        print(f"  ❌ '{query.title()}' is NOT available in stock.")
        print("  💡 Please check with the shopkeeper for restock date.")
````

  📍 Shelf Index: 1 (Row 1, Slot 2)
