# 🌾 Question 4: Ration Eligibility Calculator

**Concepts Covered:** Conditional logic, input handling, decision trees

**Scenario:** During ration distribution drives, family size determines the quantity of goods a household receives. Uncle needs a quick eligibility checker.

````python
# ============================================================
#  Kirana Store - Ration Eligibility Calculator
#  Assignment 1 | Question 4
# ============================================================

print("=" * 45)
print("   🌾 RATION ELIGIBILITY CALCULATOR")
print("=" * 45)

# --- Input ---
name    = input("Enter family head's name : ").strip().title()
members = int(input("Enter number of family members: "))

if members < 1:
    print("\n⚠️  Invalid input! Number of members must be at least 1.")
else:
    # --- Eligibility Decision Tree ---
    print("\n" + "=" * 45)
    print(f"   📋 RATION SLIP FOR: {name}")
    print("=" * 45)
    print(f"   Family Members  : {members}")
    print("-" * 45)

    if 1 <= members <= 2:
        category = "Category A (Small Family)"
        ration   = {
            "Rice"  : "10 kg",
            "Sugar" : "5 kg"
        }
    elif 3 <= members <= 5:
        category = "Category B (Medium Family)"
        ration   = {
            "Rice"  : "20 kg",
            "Sugar" : "10 kg",
            "Oil"   : "2 L"
        }
    else:
        category = "Category C (Large Family)"
        ration   = {
            "Rice"  : "35 kg",
            "Sugar" : "15 kg",
            "Oil"   : "5 L"
        }

    print(f"   Eligibility     : {category}")
    print("-" * 45)
    print("   📦 Items Eligible:")
    for item, qty in ration.items():
        print(f"      ✅ {item:<10}: {qty}")

    print("=" * 45)
    print("   Please collect your ration within 7 days.")
    print("   🙏 Jai Hind!")
    print("=" * 45)
````
