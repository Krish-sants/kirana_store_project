# 🪔 Question 5: Pattern Printing for Kirana Festive Board

**Concepts Covered:** Loops, pattern printing, string repetition

**Scenario:** Uncle wants to decorate the shop for Diwali with a chalkboard showing the store's name in a centered pyramid format.

````python
# ============================================================
#  Kirana Store - Festive Pattern Printer
#  Assignment 1 | Question 5
# ============================================================

print("=" * 50)
print("   🪔 KIRANA STORE - DIWALI FESTIVE BOARD")
print("=" * 50)

# --- Input ---
store_name = input("Enter your store name (or press Enter for 'KIRANA'): ").strip().upper()
if store_name == "":
    store_name = "KIRANA"   # Default

n = int(input("Enter number of rows for the pyramid: "))

print("\n" + "=" * 50)
print(f"   ✨ {store_name} FESTIVE PYRAMID ✨")
print("=" * 50)

# --- Pattern Logic ---
# Max width = n words + (n-1) spaces between them
word_len   = len(store_name)
max_width  = n * word_len + (n - 1) * 1   # 1 space between words

for row in range(1, n + 1):
    # Build row content: words separated by a space
    row_content = " ".join([store_name] * row)
    # Center the content
    print(row_content.center(max_width + 4))

print("=" * 50)
print(f"   Happy Diwali from {store_name}! 🪔🎆")
print("=" * 50)
````
